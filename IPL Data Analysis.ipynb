# Q11. Plot a histogram of number of matches played in a season by players.

```python
plt.figure(figsize=(10,6))
plt.hist(ipl['Mat'],10,color='b',edgecolor='black')
plt.xlabel('Matches')
plt.ylabel('No. Of Players')
plt.title('Matches Played By Players',fontsize=15)
plt.show()
```

# Q12. Plot the histogram of balls faced by players.

```python
plt.figure(figsize=(10,6))
plt.hist(ipl['BF'],10,color='m',edgecolor='black')
plt.xlabel('Balls Faced')
plt.ylabel('No Of Players')
plt.title('Balls faced by Players',fontsize=20)
plt.show()
```

# Q13. Top 20 players with most runs in a season.

```python
ipl.sort_values('Runs',ascending=False)[['Player','Runs']].head(20)
```

# Q14. Bar chart for top 20 players with best strike rate in a season.

```python
plt.figure(figsize=(15,8))
Top_20_Players=ipl.sort_values('SR',ascending=False).head(20)
plt.bar(Top_20_Players['Player'],Top_20_Players['SR'])
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=90)
plt.title('Top 20 Players By Strike Rate',fontsize=20)
plt.show()
```

# Q15. Top 30 players with highest number of sixes.

```python
ipl.sort_values('6s',ascending=False)[['Player','6s']].head(30)
```

# Q16. Scatter plot of runs scored by a player v/s balls faced in a season.

```python
plt.figure(figsize=(10,6))
plt.scatter(ipl['Runs'],ipl['BF'],marker='*',color='b')
plt.xlabel('Runs By Players')
plt.ylabel('Balls Faced By Players')
plt.title('Runs scored by a player v/s Balls faced',fontsize=20)
plt.show()
```

# Q17. KDE plot for averages on a season.

```python
sns.kdeplot(ipl['Avg'],fill='True',color='m')
plt.xlabel('Average Of Players')
plt.ylabel('Density')
plt.title('KDE plot for Averages',fontsize=20)
plt.show()
```
